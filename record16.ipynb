```python
#Create a Bank account with members account number, name, type of account and balance.
#Write constructor and methods to deposit at the bank and withdraw an amount from the bank.
class Bank:
    def __init__(self,accno,name,acctype,bal):
        self.accno=accno
        self.name=name
        self.acctype=acctype
        self.bal=bal
    def deposit(self,amt):
        self.bal+=amt
        print("Amount Deposited :",amt)
        print("Balance          :",self.bal)
    def withdraw(self,amt):
        if amt<=self.bal:
            self.bal-=amt
            print("Amount Withdrawn:",amt)
            print("Balance         :",self.bal)
        else:
            print("Insufficient Balance!")
    def balence(self):
        print("Account Balance : ",self.bal)
    def display(self):
        print("Accuunt Number                : ",self.accno)
        print("Account Holder Name           : ",self.name)
        print("Account Type (Savings/Current): ",self.acctype)
        print("Account Balance               : ",self.bal)
#V
acct_no=input("Accuunt Number                : ")
c_name=input("Account Holder Name           : ")
acct_typ=input("Account Type (Savings/Current): ")
bal=0
#OBJECT
b=Bank(acct_no,c_name,acct_typ,bal)
while True:
    print("\nSelect choice")
    print("1. DEPOSITE")
    print("2. WITHDRAWELS")
    print("3. ACCOUNT DETAILS")
    print("4. BALENCE ENQUIRY")
    print("5. EXIT\n")
    ch=int(input("Enter choice    : "))
    if ch==1:
        b.deposit(float(input("Amount           : ")))
    elif ch==2:
        b.withdraw(float(input("Amount          : ")))
    elif ch==3:
        b.display()
    elif ch==4:
        b.balence()
    elif ch==5:
        break
    else:
        print("Invalid choice!")
```
